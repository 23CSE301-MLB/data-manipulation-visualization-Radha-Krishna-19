```python

import pandas as pd
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv('iris.data.txt', names=columns)
df.head()
```

```python
print(df.head(10))
print(df.tail(5))
```

```python
print(df.describe())
print(df.dtypes)
```

```python
filtered_df = df[df['sepal_length'] > 6.0]
filtered_df.head()
```

```python
df['petal_area'] = df['petal_length'] * df['petal_width']
df[['petal_length', 'petal_width', 'petal_area']].head()
```

```python
grouped = df.groupby('class')['petal_area'].mean()
print(grouped)
```

```python
df_sorted = df.sort_values(by='sepal_width', ascending=False)
df_sorted.head()
```

```python
df_with_nan = df.copy()
df_with_nan.loc[0, 'sepal_length'] = None

df_filled = df_with_nan.fillna(df_with_nan['sepal_length'].mean())
df_filled.head()
```

```python
df.to_csv('iris_processed.csv', index=False)
print("Exported as 'iris_processed.csv'")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('iris_processed.csv')
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='sepal_length', fill=True, color='skyblue')
plt.title("KDE of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='class', y='petal_length', hue='class', palette='pastel', legend=False)
plt.title("Boxplot of Petal Length by Class")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='class', palette='deep')
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df.drop(columns=['class']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue='class', palette='husl')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```
